```python
import torch
import torch.nn as nn
if torch.cuda.is_available():
    device = torch.device('cuda')
    print("Cuda available")
else:
    device = torch.device('cpu')
    print("Cuda not available")
```

```python
#creat dataset
#creat basic class
# 5 input features
# 1 output featres
# 100 sample
# inbuilt loss function
# inbuilt optimizer
# loss
#accuracy
#dataset
x_train = torch.rand(100,6)
y_train = torch.randint(0,2,(100,))
class NeuralNetwork(nn.Module):
  def __init__(self,input_features,hidden_node,output_node):
    super().__init__()
    self.Sequential  = nn.Sequential(
        nn.Linear(input_features,hidden_node),
        nn.ReLU(),
        nn.Linear(hidden_node,output_node),
        nn.Sigmoid())
  def forward(self,features):
    return self.Sequential(features)
Model = NeuralNetwork(input_features=x_train.shape[1],hidden_node=3,output_node=1)
loss_function = nn.BCELoss()
optimizer = torch.optim.Adam(Model.parameters(),lr=0.01)
epochs=1000
for epoch in range(epochs):
  y_pred = Model(x_train)
  loss = loss_function(y_pred,y_train.reshape(-1,1).float())
  optimizer.zero_grad() #after epoch gradient will be zero it will same for each epoch
  loss.backward()      # calculate gradient
  optimizer.step()  #to update parameters
  #calculate loss
  print(f"epoch {epoch+1} loss {loss.item()}")
  with torch.no_grad(): #it will never calculate the gradient while checking the accuracy
    y_pred = Model.forward(x_train)
    y_pred = (y_pred > 0.5).float()
    accuracy = (y_pred == y_train).float().mean()
    print(f"epoch {epoch+1} accuracy {accuracy.item()}")
```
